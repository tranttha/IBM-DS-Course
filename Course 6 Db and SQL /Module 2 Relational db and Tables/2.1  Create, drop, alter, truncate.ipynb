1. Create a new table in the database named Toys with attributes as ID (integer), Variety (variable length string), and Quantity (integer). Make sure the ID is not Null.

```python
import mysql.connector as sqlconnector
conn = sqlconnector.connect(
    host='localhost',
    user='root',
    password='my-secret-pw',
    database='testdb',
    port=3306
)
```

```python
# conn = sqlconnector.connect('Toys.db')
cursor = conn.cursor()
```

```python
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Toys (
    ID INT NOT NULL,
    Variety VARCHAR(20),
    Quantity INT
    )
    '''
    )
```

2. Add the below-mentioned entries to the table using the INSERT statement.

| ID	| Variety	| Quantity |
| ---- | ----- | -------|
| 1| 	Chew toy| 	20| 
| 2	| Balls	| 50| 
| 3| 	Bowls| 	30|
|4 |	Foldable bed |	40 |


```python
# cursor = conn.cursor()
cursor.execute(''' 
    INSERT INTO Toys VALUES 
    ( 1, 'Chew toy', 20),
    (2, 'Balls', 50),
    (3, 'Bowls', 30),
    (4, 'Foldable bed', 40)
    '''   
    )
```


3. ALTER the length of 'Variety' in the table to 30 characters.


```python
# cursor = conn.cursor()
cursor.execute(
    '''
    ALTER TABLE Toys
    MODIFY Variety VARCHAR(30);
    '''
)
```


4. TRUNCATE the table 'Toys'


```python
cursor.execute(
    '''
    TRUNCATE TABLE Toys;
    ''')
```


5. DROP the table 'Toys'


```python
cursor.execute('''
    DROP TABLE Toys;
    '''
    )
```
