## Create your first R script file
1. from the menu click File > New File > R Script
2. File > Save, and name the file "...". After the script file is saved, you can see an empty file called "..."
3. click first_script.R file to add the code 
**You need to make sure the last line of the file is a new empty line.**
4. Now, you can run the code in the script file, there are two running modes:
- *Run the current line or selection*. You can click and drag your mouse or use Shift + Up/Down keys to select all lines and then click the following Run icon to run them
- The second mode is called *Source* where it runs all lines of code in the file by clicking the following Source icon
