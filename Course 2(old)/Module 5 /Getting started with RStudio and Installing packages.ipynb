### Step 2 - Now you load the iris dataset. 
Enter the following lines into the editor window that appears. Then select all the text, and click Run just above the editor window.

```r
library (datasets) ## test dataset to practice on 
data(iris)
# View(iris) ## not supported in Jupyter R kernel
head(iris)
```

You are taken to the data view tab to inspect your dataset. The dataset contains five columns and the first four are floating point type while the last column is a label of data type string which contains the category value. You can see there are total 150 entries of which you can see the first 7.
